```python
import pandas as pd
import geopandas as gpd
import plotly.express as px

# Read the data



# Ensure that 'COD_UBIGEO' and 'ubigeo' are strings with leading zeros
data['COD_UBIGEO'] = data['COD_UBIGEO'].str.zfill(6)
gdf['ubigeo'] = gdf['ubigeo'].str.zfill(6)

# Merge the data with the geodataframe
merged = gdf.merge(data, left_on='ubigeo', right_on='COD_UBIGEO', how='left')

# Prepare the GeoJSON with 'id' properties matching 'ubigeo'
def create_geojson(df, id_field):
    geojson = {'type': 'FeatureCollection', 'features': []}
    for _, row in df.iterrows():
        if row['geometry'] is not None:
            feature = {
                'type': 'Feature',
                'geometry': row['geometry'].__geo_interface__,
                'properties': {},
                'id': row[id_field]
            }
            geojson['features'].append(feature)
    return geojson

geojson_data = create_geojson(merged, 'ubigeo')

# Convert 'SIEMBRA' to numeric, handling missing values
merged['SIEMBRA'] = pd.to_numeric(merged['SIEMBRA'], errors='coerce')

# Function to plot the map for a given year and product
def plot_map(year, producto):
    # Filter the data
    df_filtered = merged[(merged['AÑO'] == year) & (merged['CROP_BASED'] == producto)]

    # Remove rows with missing 'SIEMBRA' values
    df_filtered = df_filtered.dropna(subset=['SIEMBRA'])

    # Create the map
    fig = px.choropleth_mapbox(
        df_filtered,
        geojson=geojson_data,
        locations='ubigeo',
        featureidkey='id',
        color='SIEMBRA',
        color_continuous_scale="YlOrRd",
        mapbox_style='carto-positron',
        zoom=4.5,
        center={"lat": -9.19, "lon": -75.0152},
        opacity=0.6,
        labels={'SIEMBRA': 'Sowing (hectares)'},
        hover_data={
            'DEPARTAMENTO': True,
            'PROVINCIA': True,
            'DISTRITO': True,
            'SIEMBRA': ':.2f'
        }
    )
    fig.update_layout(
        title_text=f"Sowing Data for {producto} in {year}",
        margin={"r":0,"t":30,"l":0,"b":0}
    )
    fig.show()

# Example usage
plot_map(2015, 'PAPA')
```

```python
import pandas as pd
import geopandas as gpd
import plotly.express as px
import unicodedata
import json
import nbformat

# Function to normalize department names
def normalize_name(name):
    name = name.upper()
    name = unicodedata.normalize('NFKD', name).encode('ASCII', 'ignore').decode('ASCII')
    name = name.strip()
    return name

# Read the data
# Replace 'data.csv' with the path to your data file
# Convert 'SIEMBRA' to numeric, handling decimal commas
data['SIEMBRA'] = data['SIEMBRA'].astype(float)

# Normalize department names in the data
data['DEPARTAMENTO_NORM'] = data['DEPARTAMENTO'].apply(normalize_name)

# Aggregate SIEMBRA per DEPARTAMENTO per MES in a given year
aggregated_data = data.groupby(['AÑO', 'MES', 'DEPARTAMENTO_NORM'])['SIEMBRA'].sum().reset_index()

# Normalize department names in the GeoDataFrame
gdf['NOMBDEP_NORM'] = gdf['NOMBDEP'].apply(normalize_name)

# Convert the GeoDataFrame to GeoJSON
geojson_data = json.loads(gdf.to_json())

# Function to plot the map for a given year and month
def plot_map(year, month):
    # Filter the data for the given year and month
    df_filtered = aggregated_data[(aggregated_data['AÑO'] == year) & (aggregated_data['MES'] == month)]
    
    # Remove rows with missing 'SIEMBRA' values
    df_filtered = df_filtered.dropna(subset=['SIEMBRA'])
    
    # Merge with GeoDataFrame to ensure matching
    df_plot = df_filtered.merge(gdf[['NOMBDEP_NORM']], left_on='DEPARTAMENTO_NORM', right_on='NOMBDEP_NORM', how='left')
    
    # Create the map
    fig = px.choropleth_mapbox(
        df_plot,
        geojson=geojson_data,
        locations='DEPARTAMENTO_NORM',
        featureidkey='properties.NOMBDEP_NORM',
        color='SIEMBRA',
        color_continuous_scale="YlOrRd",
        mapbox_style='carto-positron',
        zoom=4.5,
        center={"lat": -9.19, "lon": -75.0152},
        opacity=0.6,
        labels={'SIEMBRA': 'Sowing (hectares)'},
        hover_name='DEPARTAMENTO_NORM',
        hover_data={'SIEMBRA': ':.2f'}
    )
    fig.update_layout(
        title_text=f"Sowing Data for {year}-{str(month).zfill(2)}",
        margin={"r":0,"t":30,"l":0,"b":0}
    )
    fig.show()

# Example usage
plot_map(2015, 3)  # Plot for March 2015
```
