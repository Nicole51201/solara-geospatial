```python
# Diccionario de agrupación de hortalizas
hortalizas_agrupadas = {
    'RABANO': 'RABANO',
    'ZANAHORIA': 'ZANAHORIA',
    'COL O REPOLLO': 'COL/REPOLLO',
    'COL': 'COL/REPOLLO',
    'ESPINACA': 'ESPINACA',
    'LECHUGA': 'LECHUGA',
    'LECHUGA VERDE': 'LECHUGA',
    'LECHUGA NEGRA': 'LECHUGA',
    'CEBOLLA CHINA': 'CEBOLLA CHINA',
    'PORO': 'PORO',
    'BROCOLI': 'BROCOLI',
    'AJI ESCABECHE': 'AJI',
    'AJI AMARILLO': 'AJI',
    'AJI PANCA': 'AJI',
    'AJI TABASCO': 'AJI',
    'AJI LIMO': 'AJI',
    'AJI CHALLUARURO': 'AJI',
    'AJI MONTANA': 'AJI',
    'AJI CRIOLLO': 'AJI',
    'PEPINO (SOLANUM MURICATUM)': 'PEPINO',
    'PEPINILLO': 'PEPINO',
    'PIMIENTO': 'PIMIENTO',
    'ESPARRAGO': 'ESPARRAGO',
    'ACELGA': 'ACELGA',
    'COLIFLOR': 'COLIFLOR',
    'NABO': 'NABO',
    'PEREJIL': 'PEREJIL',
    'CEBOLLA': 'CEBOLLA',
    'CEBOLLA CABEZA AMARILLA': 'CEBOLLA CABEZA',
    'CEBOLLA CABEZA ROJA': 'CEBOLLA CABEZA',
    'ALBAHACA': 'ALBAHACA',
    'ALBAHACA (ESPECIAL)': 'ALBAHACA',
    'ZAPALLO ITALIANO': 'ZAPALLO ITALIANO',
    'TOMATE': 'TOMATE',
    'BERENJENA': 'BERENJENA',
    'RUDA': 'RUDA',
    'CALIFORNIA WONDER': 'PIMIENTO',
    'FRIJOL VAINITA': 'FRIJOL VAINITA',
    'COL CHINA': 'COL CHINA'
}

# Función para agrupar los nombres
def agrupar_hortalizas(cultivo):
    cultivo_normalizado = cultivo.upper().strip()  # Normalizar el nombre
    return hortalizas_agrupadas.get(cultivo_normalizado, None)  # Devolver el nombre agrupado o None si no es hortaliza
```

```python
tudo['hortaliza_agrupada'] = tudo['PRODUCTO'].apply(agrupar_hortalizas)
```

```python
# Filtrar solo las hortalizas
df_hortalizas = tudo[tudo['hortaliza_agrupada'].notnull()]
```
