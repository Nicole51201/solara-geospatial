```python
import requests
from bs4 import BeautifulSoup
import csv


# URL of the page
url = "http://sistemas.midagri.gob.pe/sisap/portal2/mayorista/"
# Make a GET request to fetch the page content
response = requests.get(url)
# Check if request was successful
if response.status_code == 200:
    content = response.text

    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(content, "html.parser")

    # Find the div containing all product list items
    productos_div = soup.find("div", {"id": "productos"})
    print(productos_div)
    # Initialize a list to store product codes and names
    product_list = []

    # Check if the productosCheckBox div was found
    if productos_div:
        print('gaa')
        # Iterate over each <li> element within the div
        for li in productos_div.find_all("li"):
            # Find the checkbox within the <li> to get the code
            checkbox = li.find("input", {"type": "checkbox", "name": "productos[]"})
            if checkbox:
                product_code = checkbox.get("value")  # Get the product code
                
                # Find the label containing the product name as plain text
                label_tag = li.find("label")
                if label_tag:
                    # Get the text directly from the label and strip whitespace
                    product_name = label_tag.get_text(strip=True).split()[-1]  # Extracting the last part, assuming it’s the name
                    
                    # Append the code and name to the product list
                    product_list.append({"code": product_code, "name": product_name})
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
data = pd.DataFrame(product_list)
data['code'] = data['code'].astype(str)
data.to_csv('Cultivos_Codigos_SISAP.csv',index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import time

# Load the codigos_productos DataFrame
# Assuming 'codigos_productos.csv' is the file with product codes and names

# Configure the Selenium WebDriver (using Chrome here)
driver = webdriver.Chrome()

# Dictionary to store data for each crop
crop_data2 = {}
# List to store problematic products
problematic_products = []

# Open the page once before iterating through crops
driver.get("http://sistemas.midagri.gob.pe/sisap/portal2/mayorista/")

# Iterate over each crop in the codigos_productos DataFrame
for index, row in remaining_codigos_productos.iterrows():
    product_code = row['code']
    product_name = row['name']
    
    try:
        # Refresh page to reset state for each crop
        driver.refresh()
        time.sleep(1)
        
        # Wait for and click the "intervalo" button
        intervalo_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.ID, "intervalo"))
        )
        intervalo_button.click()

        # Set date range
        fecha_hasta = datetime.now()
        fecha_desde_str = '01/01/2020'
        fecha_hasta_str = fecha_hasta.strftime("%d/%m/%Y")
        
        # Set date range in input fields
        desde_input = driver.find_element(By.ID, "desde")
        desde_input.clear()
        desde_input.send_keys(fecha_desde_str)

        hasta_input = driver.find_element(By.ID, "hasta")
        hasta_input.clear()
        hasta_input.send_keys(fecha_hasta_str)
        
        # Select the crop by replacing the product code in the XPath
        cultivo_xpath = f"//input[@value='{product_code}']"
        cultivo = driver.find_element(By.XPATH, cultivo_xpath)
        cultivo.click()

        # Click "Consultar"
        time.sleep(2)
        consultar_button = driver.find_element(By.ID, "consultar")
        consultar_button.click()

        # Wait for either the table content or error message to load
        WebDriverWait(driver, 30).until(
            EC.visibility_of_element_located((By.CSS_SELECTOR, "#reporte .contenido"))
        )

        # Re-fetch page source to include loaded data
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, "html.parser")

        # Check for the error message
        error_message = soup.find("p", class_="mensajeDeError")
        if error_message and "Demasiados criterios" in error_message.text:
            print(f"Skipping {product_name} due to too many criteria error.")
            problematic_products.append(product_name)
            continue  # Skip to the next product

        # Process the table if no error
        table_div = soup.find("div", {"id": "reporte"})

        if table_div:
            table = table_div.find("table")
            
            if table:
                data = []
                
                # Extract headers (if any), using both 'td' and 'th'
                headers = [header.text.strip() for header in table.find("tr", {"class": "encabezado"}).find_all(["td", "th"])]
                
                # Extract rows with data (class 'contenido')
                for row in table.find_all("tr", {"class": "contenido"}):
                    columns = row.find_all("td")
                    data.append([col.text.strip() for col in columns])

                # Create DataFrame if headers and data are found
                if headers and data:
                    crop_df = pd.DataFrame(data, columns=headers)
                    
                    # Store the DataFrame in the dictionary
                    crop_data2[product_name] = crop_df
                    print(f"Data for {product_name} collected successfully.")
                else:
                    print(f"No valid data for {product_name}.")
            else:
                print(f"No table found for {product_name}.")
        else:
            print(f"No 'reporte' div found for {product_name}.")

        # Pause between requests to avoid overwhelming the server
        time.sleep(2)
    
    except Exception as e:
        print(f"An error occurred with {product_name}: {e}")
        problematic_products.append(product_name)
        continue  # Skip to the next product

# Close the driver after all crops have been processed
driver.quit()

# Print problematic products for review
print("Problematic products:")
print(problematic_products)

# Optionally, save problematic products to a file
with open('problematic_products.txt', 'w') as f:
    for item in problematic_products:
        f.write("%s\n" % item)
```
